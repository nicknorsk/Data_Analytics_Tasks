```python
# Step 1. Create data list
data = [1, 3, 4]

# Step 2. Calculate Σx (sum of all x)
sum_x = sum(data)
print(f"Σx = {sum_x}")

# Step 3. Calculate Σx² (sum of squares)
sum_x_squared = sum([x**2 for x in data])
print(f"Σx² = {sum_x_squared}")

# Step 4. Calculate Σ(x-1)² (sum of squared differences)
sum_x_minus_1_squared = sum([(x-1)**2 for x in data])
print(f"Σ(x-1)² = {sum_x_minus_1_squared}")

# Σx = 1 + 3 + 4 = 8
# Σx² = 1² + 3² + 4² = 1 + 9 + 16 = 26
# Σ(x-1)² = (1-1)² + (3-1)² + (4-1)² = 0 + 4 + 6 = 10
```

  3.1. без використання жодних бібліотек, користуючись тільки Python.  
  3.2. використовуючи відповідний метод бібліотеки numpy.  

Виведіть результат обчислень в кожному з випадків на екран. Порівняйте результати зі значенням вибіркового середнього. Напишіть коментар, для яких задач ви б використовували вибіркове середнє, а для яких - медіану?

```python
# 3.1 Pure Python method
def calculate_median_python(data):
    # Sort the data
    sorted_data = sorted(data)
    n = len(sorted_data)
    # Find the middle position
    mid = n // 2

    # If even number of elements
    if n % 2 == 0:
        median = (sorted_data[mid-1] + sorted_data[mid]) / 2
    # If odd number of elements
    else:
        median = sorted_data[mid]
    return median

python_median = calculate_median_python(sum_purchases)
print(f"Method 3.1 - Median (Pure Python): {python_median:.2f} UAH")

# 3.2 NumPy method
import numpy as np
numpy_median = np.median(sum_purchases)
print(f"Method 3.2 - Median (NumPy): {numpy_median:.2f} UAH")

# Compare with mean
mean_value = np.mean(sum_purchases)
print(f"\nMean value: {mean_value:.2f} UAH")

# Print analysis
print("\nAnalysis:")
print("1. The median (319.00) is much lower than the mean (667.95)")
print("2. This indicates skewed data with outliers")
print("3. Median is more representative here because:")
print("   - It's not affected by extreme values (4310 and 3524 UAH)")
print("   - It better represents the 'typical' purchase amount")
```

**When to use each measure:**

**Mean:**

1. For symmetric data distributions
2. When outliers are valid and should be included
3. For further statistical calculations

**Median:**

1. For skewed distributions (like our data)
2. When there are outliers that could distort the average
3. For income/price data analysis
4. When looking for a "typical" value

In this case, the median is more appropriate as it gives a better picture of the typical purchase amount, unaffected by the two unusually large purchases.

**Завдання 4**. Обчислість стандартне відхилення наведеної вибірки в змінній `sum_purchases` кожним з зазначених методів:  

  4.1. без використання жодних бібліотек, користуючись тільки Python.  
  4.2. використовуючи відповідний метод бібліотеки numpy.  

Виведіть результат обчислень в кожному з випадків на екран.

```python
# 4.1 Pure Python method
def calculate_std_python(data):
    n = len(data)
    # Calculate mean
    mean = sum(data) / n
    # Calculate sum of squared differences
    squared_diff_sum = sum((x - mean) ** 2 for x in data)
    # Calculate variance and standard deviation
    variance = squared_diff_sum / (n - 1)  # Using n-1 for sample standard deviation
    std_dev = variance ** 0.5
    return std_dev

# 4.2 NumPy method
import numpy as np

# Calculate both standard deviations
python_std = calculate_std_python(sum_purchases)
numpy_std = np.std(sum_purchases, ddof=1)  # ddof=1 for sample standard deviation

print(f"Method 4.1 - Standard Deviation (Pure Python): {python_std:.2f} UAH")
print(f"Method 4.2 - Standard Deviation (NumPy): {numpy_std:.2f} UAH")
```

```python
# Compare standard deviation with different ddof values
std_ddof0 = np.std(sum_purchases, ddof=0)  # Population standard deviation
std_ddof1 = np.std(sum_purchases, ddof=1)  # Sample standard deviation

print("Understanding ddof (Delta Degrees of Freedom):")
print(f"Standard Deviation with ddof=0: {std_ddof0:.2f}")
print(f"Standard Deviation with ddof=1: {std_ddof1:.2f}")

# Show the formulas in action
n = len(sum_purchases)
mean = np.mean(sum_purchases)
squared_diff_sum = sum((x - mean) ** 2 for x in sum_purchases)

# Calculate both versions
pop_std = (squared_diff_sum / n) ** 0.5        # ddof=0
sample_std = (squared_diff_sum / (n-1)) ** 0.5 # ddof=1

print("\nCalculated manually:")
print(f"Population std (n):   {pop_std:.2f}")
print(f"Sample std (n-1):     {sample_std:.2f}")
```

**Завдання 7**. В попередньому завданні ми помітили, що викиди нам заважають отримати "правдиве" значення кореляції. Давайте вилучимо записи, які відповідають значенням з викидами в сумі покупок і обчисліть коеф. кореляції. Для цього поставте `threshold`, який дорінює середньому значенню + 2 стандартних вдіхилля суми покупок і всі значення, які більше за нього, будемо вважати викидами (є і інші правила для визначення викидів - з ним и познайомимось пізніше).

Вилучіть значення з обох масивів і обчисліть коеф. кореляції Пірсона з допомогою `numpy` для кількості візитів і суми покупок без записів-викидів. Виведіть результат на екран, побудуйте діаграму розсіювання і проаналізуйте результат: що можемо сказати про залежність в даних?

```python
import numpy as np
import matplotlib.pyplot as plt

sum_purchases = [316, 493, 367, 153, 224, 121, 322, 450, 196, 397, 281,
                 239, 484, 124, 223, 448, 432, 255, 4310, 3524]
visits = [6, 2, 10, 3, 9, 9, 2, 3, 6, 6, 10, 2, 5, 10, 3, 5, 4,
          7, 4, 2]

# Calculate threshold
mean = np.mean(sum_purchases)
std = np.std(sum_purchases)
threshold = mean + 2 * std

print(f"Mean: {mean:.2f}")
print(f"Standard deviation: {std:.2f}")
print(f"Threshold: {threshold:.2f}")

# Remove outliers
filtered_purchases = []
filtered_visits = []
outliers_purchases = []
outliers_visits = []

for purchase, visit in zip(sum_purchases, visits):
    if purchase <= threshold:
        filtered_purchases.append(purchase)
        filtered_visits.append(visit)
    else:
        outliers_purchases.append(purchase)
        outliers_visits.append(visit)

# Calculate correlation for filtered data
correlation = np.corrcoef(filtered_purchases, filtered_visits)[0,1]

print(f"\nNumber of total points: {len(sum_purchases)}")
print(f"Number of points after filtering: {len(filtered_purchases)}")
print(f"Number of outliers removed: {len(outliers_purchases)}")
print(f"\nCorrelation coefficient after removing outliers: {correlation:.4f}")

# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(filtered_purchases, filtered_visits, label='Normal values', color='blue')
plt.scatter(outliers_purchases, outliers_visits, label='Outliers', color='red')

plt.axvline(x=threshold, color='r', linestyle='--', label='Threshold')

plt.title('Scatter Plot: Purchases vs Visits (Outliers Highlighted)')
plt.xlabel('Sum of Purchases (UAH)')
plt.ylabel('Number of Visits')
plt.legend()
plt.grid(True)
plt.show()
```

**Conclusions:**

1. The relationship between visits and purchases is more apparent after removing outliers
2. The correlation coefficient provides a better representation of the typical customer behavior
3. Outliers should be analyzed separately as they might represent special cases or different customer segments
