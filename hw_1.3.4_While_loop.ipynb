```python
limit = 9  # Задане число
```

```python
# Step 1. Create a counter:

i = 1

# Step 2. Create while loop to count from 1 to limit:
while i <= limit:
  print(i)
  i += 1

# print the end message:
print("The end of calculation!")
```


3. **Обчислення факторіалу.**   
  Факторіал числа - це добуток всіх натуральних чисел від 1 до цього числа включно. Наприклад, факторіал числа 5 (позначається як 5!) дорівнює `1 * 2 * 3 * 4 * 5 = 120`.

  Завдання:  
      - Обчисліть факторіал заданого числа, використовуючи цикл `while`.  
      - Задайте деяке число для обчислення факторіалу у змінній `num`.  
      - Виведіть обчислений факторіал на екран.  


```python
# Step 1. Initialize number to calculate the factorial in the variable:
num = 9

# Step 2. Factorial:
factorial = 1

# Step 3. Counter for the while loop:
counter = factorial

# Step 4. Calculate factorial using while loop:
while counter <= num:
  factorial *= counter
  counter += 1
print(factorial)

# Step 5. Check of actual result with the multiplication operation by math formula:
n9 = 1 * 2 * 3 * 4 * 5 * 6 * 7 * 8 * 9
n9
```
