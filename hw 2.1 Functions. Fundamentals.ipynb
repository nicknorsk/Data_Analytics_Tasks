````python
# Solution 1. The function outputs text by using "for" loop:
def print_message():
  for i in range (5):
    print("–Ø –ø—Ä–æ–≥—Ä–∞–º—É—é –Ω–∞ Pythonüí™")

print_message()
````

````python
# Solution 2. The function "print_message" outputs text by using "while" loop:

def print_message():
  i = 1
  while i <= 5:
    print("–Ø –ø—Ä–æ–≥—Ä–∞–º—É—é –Ω–∞ Pythonüí™")
    i += 1

print_message()
````

````python
# Solution 3. Create function and "for" loop

def multiply_numbers(num1, num2):
   """
   Multiplies two numbers and returns their product.
   
   :param num1: First number to multiply
   :type num1: int or float
   :param num2: Second number to multiply
   :type num2: int or float
   :returns: Product of the two input numbers
   :rtype: int or float
   """
   return num1 * num2

# List of number pairs to multiply

number_pairs = [
   (10, 2),
   (7, 8),
   (1024, 999)
]

# Loop through pairs and print results

for num1, num2 in number_pairs:
   result = multiply_numbers(num1, num2)
   print(f"{num1} * {num2} = {result}")
````

````python
# Solution 3.

def int_division(a, b):
    # Check if both arguments are integers
    if not isinstance(a, int) or not isinstance(b, int):
        return "–ù–µ –ø—Ä–∞—Ü—é—é –∑ —Ç–∞–∫–∏–º–∏ —Ç–∏–ø–∞–º–∏ –¥–∞–Ω–∏—Ö."

    # Check if the second argument is zero
    if b == 0:
        return "–ù–µ –º–æ–∂—É –≤–∏–∫–æ–Ω–∞—Ç–∏ –¥—ñ–ª–µ–Ω–Ω—è –Ω–∞ 0."

    # Perform division
    return a / b

# Test cases and outputs
print(f"Function result: {int_division(12, 6)}.")
print(f"Function result: {int_division(20, 0)}.")
print(f"Function result: {int_division(112, 10)}.")
print(f"Function result: {int_division(99, 0)}.")
print(f"Function result: {int_division(11, 1.1)}.")
````
