
2. Використовуючи фукнціонал `numpy` створіть масив `x` з 400 точок в діапазоні між -2 та 2 і масив `x_positive` в діапазоні між 0.1 і 2 також з 400 точок. Другий масив в нас буде для логарифмічної функції, перший з більш широким діапазоном - для інших функцій.


```python
# Step 1. Create 1st array "x"
x = np.linspace(-2, 2, 400)
x
```

```python
# Step 2. Create 2nd array "x_positive" for logarithmic function
x_positive = np.linspace(0.1, 2, 400)
x_positive
```

```python
# Step 3. Verification of arrays created:
print("x array:")
print(f"Number of points: {len(x)}")
print(f"Range: from {x[0]} to {x[-1]}")
print("\nx_positive array:")
print(f"Number of points: {len(x_positive)}")
print(f"Range: from {x_positive[0]} to {x_positive[-1]}")
```


3. Тут маємо кілька кроків.

  3.1. Ініціалізуйте фігуру Matplotlib з розміром, достатнім для трьох панелей (рекомендую спробувати експериментально виявити добрий розмір).   
  3.2. На першій панелі візуалізуйте експоненціальну функцію `y = e^x`. Використовуйте відповідну змінну `x`.  
  3.2. На другій панелі покажіть логарифмічну функцію `y = ln(x)`.  
  3.4. На третій панелі покажіть модифіковану синусоїду `y = sin(x) + x`.  

  Відобразіть графік.


```python
# Task 3
# Solution 1. Object oriented approach:

# Step 3.1. Create a figure:
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))

# Step 3.2. Assign the first panel: exponential function:
y_exp = np.exp(x)
ax1.plot(x, y_exp)

# Step 3.3. Assign the second panel: logarithmic function:
y_log = np.log(x_positive)
ax2.plot(x_positive, y_log)

# Step 3.4. Assign the third panel: modifies sine function
y_sin = np.sin(x) + x
ax3.plot(x, y_sin);
```

```python
# Task 3
# Solution 2. Matlab style approach:

# Step 3.1. Create figure
plt.figure(figsize=(15, 5))

# Step 3.2. First subplot: exponential function
plt.subplot(1, 3, 1)  # 1 row, 3 columns, first position
plt.plot(x, np.exp(x))

# Step 3.3. Second subplot: logarithmic function
plt.subplot(1, 3, 2)  # 1 row, 3 columns, second position
plt.plot(x_positive, np.log(x_positive))

# Step 3.4. Third subplot: modified sine function
plt.subplot(1, 3, 3)  # 1 row, 3 columns, third position
plt.plot(x, np.sin(x) + x);
```

4. Модифікація і довершення графіку.  
  4.1. Для кожної панелі встановіть назву графіка та підписи до осей х та у (можна так і назвати `x` та формула функції).  
  4.2. Використайте різні кольори для кожного графіка.  


```python
# Task 4.
# Solution 1. Matlab style approach:

# Step 1. Create the figure:
plt.figure(figsize=(15, 5))

# Step 1. First subplot: Exponential function
plt.subplot(1, 3, 1)  # 1 row, 3 columns, first position
plt.plot(x, np.exp(x), 'b-', label='y = e^x')
plt.grid(True)
plt.title('Exponential Function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Step 2. Second subplot: Logarithmic function
plt.subplot(1, 3, 2)  # 1 row, 3 columns, second position
plt.plot(x_positive, np.log(x_positive), 'r-', label='y = ln(x)')
plt.grid(True)
plt.title('Logarithmic Function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Step 3. Third subplot: Modified sine function
plt.subplot(1, 3, 3)  # 1 row, 3 columns, third position
plt.plot(x, np.sin(x) + x, 'g-', label='y = sin(x) + x')
plt.grid(True)
plt.title('Modified Sine Function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Step 4. Adjust layout and display
plt.tight_layout()
#plt.show()
```

```python
# Task 4.
# Solution 2. Object oriented approach:

# Step 1. Create a figure:
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))

# Step 2. Assign the first panel: exponential function:
y_exp = np.exp(x)
ax1.plot(x, y_exp, 'b-', label='y = e^x')
ax1.grid(True)
ax1.set_title('Exponential function')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.legend()

# Step 3. Assign the second panel: logarithmic function:
y_log = np.log(x_positive)
ax2.plot(x_positive, y_log, 'r-', label='y = ln(x)')
ax2.grid(True)
ax2.set_title('Logarithmic function')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.legend()

# Step 4. Assign the third panel: modified sine function
y_sin = np.sin(x) + x
ax3.plot(x, y_sin, 'g-', label='y = sin(x) + x')
ax3.grid(True)
ax3.set_title('Modified sine function')
ax3.set_xlabel('X')
ax3.set_ylabel('Y')
ax3.legend()

# Step 5. Adjust layout and display
plt.tight_layout()
#plt.show()
```

5. А тепер зобразіть ті самі 3 трафіки тих самих кольорів тільки на одному графіку. Визначіть назву кожної з ліній для легенди та обовʼязково відобразіть легенду.

```python
# Task 5
# Solution 1.
# Exponential function
plt.plot(x, np.exp(x), 'b-', label='y = e^x')
plt.grid(True)
plt.title('Exponential Function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Logarithmic function
plt.plot(x_positive, np.log(x_positive), 'r-', label='y = ln(x)')
plt.grid(True)
plt.title('Logarithmic Function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Modified sine function
plt.plot(x, np.sin(x) + x, 'g-', label='y = sin(x) + x')
plt.grid(True)
plt.title('Modified Sine Function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Adjust layout and display
plt.tight_layout()
#plt.show()
```
