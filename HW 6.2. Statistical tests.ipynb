1. Зчитані дані - це сума покупок в доларах наших покупців на сайті протягом липня 2022 року.

До цього місяця, ми мали продажі в середньому на чек $20. Чи вийшло у нас статистично значущо **підвищити** середній чек за останній місяць?

Оберіть статистичний тест (z-test чи t-test). Виведіть p-значення статистичного тесту та зробіть висновок, чи ми маємо зараз статистично значущу різницю у розмірі середнього чеку.

**Підказка**: тут тест буде одновибірковий.

```python
from scipy import stats

# Step 1. The known value of the average check up to this month
mu0 = 20

# Step 2. Conduct a one-sample t-test
t_stat, p_value = stats.ttest_1samp(data, mu0)

print(f'p-value: {p_value}')
print(f'Average value of the sample: {np.mean(data):.2f}')
```

Conclusion:

1. Since p-value (0.322) is greater than the significance level of 0.05, we cannot reject the null hypothesis.
2. Although we see a slight decrease in the average check from 20 USD to 19.38 USD, this difference is not statistically significant.
3. This means that any observed differences are likely due to random variation rather than a real change in customer behavior.

4. We did not achieve a statistically significant change in the average check value this month.

```python
import pandas as pd
import numpy as np
from statsmodels.stats.weightstats import ztest

# Step 1. Read CSV files
df1 = pd.read_csv('/drive/MyDrive/Colab_Notebooks/Data_Analyst_Course_HP/Module_6/Data_M6/data.csv')
df2 = pd.read_csv('/drive/MyDrive/Colab_Notebooks/Data_Analyst_Course_HP/Module_6/Data_M6/data_2.csv')

# Step 2. Convert to numpy arrays
july_data = df1['data'].to_numpy()
august_data = df2['data'].to_numpy()

# Step 3. Calculate means for both months
print(f'July mean: {np.mean(july_data):.2f}')
print(f'August mean: {np.mean(august_data):.2f}')

# Step 4. Perform two-sample z-test
z_stat, p_value = ztest(july_data, august_data)
print(f'p-value: {p_value}')

```

**Conclusion:**

Statistical test results:

p-value (0.9859) is much larger than significance level (0.05)
This strongly suggests we cannot reject the null hypothesis
There is no statistically significant difference between the means.
